```python
def text_preprocessing(text):
    text=re.sub(r'\[[0-9]*\]', ' ', text)
    text=re.sub(r'\s+', ' ', text)
    ftext=re.sub('[^a-zA-Z]', ' ', text )
    ftext=re.sub(r'\s+', ' ', ftext)
    return text,ftext
```

```python
import re
import nltk
import requests
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from collections import Counter
 
page = requests.get('https://www.nbcnews.com/news/us-news/concerned-parentscovid-vax-tktk-rcna3901') # Getting page HTML through request
html_data = BeautifulSoup(page.content, 'html')
all_text = ""
for p in html_data.find_all('p'):
    all_text += p.get_text()
print(all_text)
```

```python
text,ftext=text_preprocessing(all_text)
print(text)
```

```python
print(ftext)
```

```python
s=stopwords.words("english")
s
```

```python
sentences = nltk.sent_tokenize(text)
sentences
```

```python
word_tokens=nltk.word_tokenize(ftext)
```

```python
w_frequencies_list =[]

for w in word_tokens:
    if w in s:
        pass
    else:
        w_frequencies_list.append(w)
w_frequencies=dict(Counter(w_frequencies_list))
w_frequencies
```

```python
max_frequency=max(w_frequencies.values())
for k,v in  w_frequencies.items():
    w_frequencies[k]=v/max_frequency
print(max_frequency)
w_frequencies
```

```python
sentence_scores = dict()
for sent in sentences:
    word_tokens=nltk.word_tokenize(sent.lower())
    for w in word_tokens:
        if w in w_frequencies:
            if len(sent.split(' ')) < 30:
                if sent in sentence_scores.keys():
                    sentence_scores[sent] += w_frequencies[w]
                else:
                    sentence_scores[sent] = w_frequencies[w]
print(sentences[0])    
sentence_scores
```

```python
summary_list=[]
for w in sorted(sentence_scores, key=sentence_scores.get, reverse=True):
    summary_list.append(w)
summary=''
for i in range(7):
    summary=summary+summary_list[i]
print(summary)
```
